```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(u"I will google about facebook")
```

```python
doc.text
```

```python
doc[-1]
```

```python
doc[2].pos_
```

```python
doc[2].tag_
```

```python
spacy.explain('VB')
```

```python
for word in doc:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
# here 'left' is verb
doc2 = nlp(u"I left the room")
for word in doc2:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
# here 'left' is noun
doc3 = nlp(u"to the left of the room")
for word in doc3:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
# here "read" is singular present
doc4 = nlp(u"I read books on history")
for word in doc4:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
# here "read" is past participle
doc5 = nlp(u"I have read a book on history")
for word in doc5:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

- EP=Emission Probability
- TP=Transition Probability
- S = Start
- E = End
- N = Noun
- M = Model
- V = Verb
- HMM = Hidden Markov Model
- Problem understand POS tagging through find the POS tagging of "wiil will google campusX"

- Step in POS tagging
  - caculate emission probability
  - calculate transition probability
  - make a graph (here we see many possible path)
  - select a that path which total probability from start to end is maximum

- The time complexity of POS method is exponential hence for optimization we use viterbi algorithm.

**Viterbi algorithm (VA): It is a recursive optimal solution to the problem of estimating the state sequence of a discrete-time finite-state Markov process observed in memoryless noise. Many problems in areas such as digital communications can be cast in this form.**
